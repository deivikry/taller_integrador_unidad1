### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí
# aqui estamos poniendo acumuladores para ir sumando
suma_total =0 
datos_total=0
for dato in esperanzas_de_vida: #aqui estamos recorriendo cada indice en nuestra lista
    suma_total +=  dato#aqui agarramos especificamente el valor en cada indice de la lista y lo suma
    datos_total = datos_total +1 # aqui solamente contamos la catidad de datos que hay 
    #al estar iniciado en 0 el primer valor va a ser 1 porque: (0+1 = 1) y para el siguiente dato la variable va a valer 1 y se repite este proceso
    
promedio = suma_total/datos_total


# Promedio
print(f"suma total entre todos los datos: {suma_total}")
print (f"datos totales : {datos_total}")
print(f"el promedio es de {promedio:.2f}")


# Máximo
dato_mayor= esperanzas_de_vida[0]
for dato in esperanzas_de_vida:
    if dato > dato_mayor:
        dato_mayor = dato
print (f'el dato mayor es {dato_mayor}')
    

# Mínimo
dato_menor = esperanzas_de_vida[0] #iniciamos una variable, que contiene el primer dato de nuestra lista
for dato in esperanzas_de_vida:#para luego recorrer toda la lista y hacer una comparativa
    if dato < dato_menor: #decimos que si en el recorrido encontramos un dato menor a la variable 'dato_menor' ejecute lo de abajo
        dato_menor=dato #asignamos un nuevo valor a dato_menor el cual es. el valor de dato que es la iteracion en nuestro bucle for
print (f'el dato menor es {dato_menor}')
````

### 2.2 — Limpieza de datos

🔧 **Ejercicio 2.2** — El dataset tiene algunos valores nulos (¡datos reales siempre tienen este problema!).

3. Explica en una celda de texto: ¿Por qué usamos la media y no simplemente eliminamos esas filas?

````python
# Tu código aquí
media =dt.select_dtypes(include="number").mean()#sacamos la media a las columnas que sean tipo numericas
#aqui agarramos a nuestro dt y le aplicamos la funcion para remplzar nulos por la media y nuesto dt queda limpio de nulos
dt = dt.fillna(media)#fillna es llenar
nuevos_nulos = dt.isnull().sum()  #is null (es nulo) .sum (sumar todos los resultados de isnull)

print (nuevos_nulos)
````

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Tu código aquí
#1
esperanza_vida_promedio_region = dt.groupby('region')['esperanza_vida'].mean()
#agrupamos por region el promedio de esperanza_vida
print (f'esperanza de vida promedio por region es : {esperanza_vida_promedio_region}')

#2
gasto_salud_promedio = dt.groupby('nivel_ingresos')['gasto_salud_usd'].mean()
print (f' gasto en salud promedio por region: \n{gasto_salud_promedio}')

#3

num_pais_for_region = dt.groupby('region')['pais'].count()
print (f'numero de paises por region: \n {num_pais_for_region}')

#4
region_mayor_esperanza_vida= esperanza_vida_promedio_region.idxmax()
valor_mayor_region = esperanza_vida_promedio_region.max()
print(f'la region con la mayor esperanza de vida es: {region_mayor_esperanza_vida} con un valor de: {valor_mayor_region:.2f}')



region_menor_esperanza_vida= esperanza_vida_promedio_region.idxmin()
valor_menor_region = esperanza_vida_promedio_region.min()
print(f'la region con la menor esperanza de vida es: {region_menor_esperanza_vida} con un valor de: {valor_menor_region:.2f}')
````

````python
# Tu código aquí
#1 aplicar funcion a esperanza vida  --> tip, las funciones o procesos son al final, primero seelige a donde se va a palicar los procesos
dt['categoria_vida'] = dt['esperanza_vida'].apply(clasificar_esperanza_vida)

total_paises = dt['categoria_vida'].value_counts()
print(f' el total  de paises que caen en una categoria esz de: {total_paises}')

porcentajes = (total_paises/len(dt)*100 )#agarramos cada dato de total paises y lo multiplicamos por 100
print('los porcentajes para cada categoria')
print(porcentajes.round(2).astype(str) + '%')
````
